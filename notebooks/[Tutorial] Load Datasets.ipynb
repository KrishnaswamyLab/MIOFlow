# Jacks

```python
n_axes = 3
df = make_jacks(
    n_axes = n_axes,
    points = 1000,
    label_by = 'axis',
#     label_by = 'coord',
    n_classes = 3,
    use_neg = True,
    data_scale = 10,
    add_noise = True,
    noise_scale = 0.03,
)

fig = plt.figure(figsize=(12, 8))
if n_axes > 2:
    ax = plt.axes(projection='3d')
    ax.scatter(df.d1, df.d2, df.d3, c=df.samples)
else:
    sns.scatterplot(data=df, x='d1', y='d2', hue='samples', palette='viridis')
```
